Exercise 3: Read all product sales data and show it using a multiline plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5), dpi=100)

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

for product in products:
    plt.plot(df.index, df[product], label=f'{product.title()} Sales Data', linewidth=2)

plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.xticks(range(1, 13))
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
import matplotlib.pyplot as plt

data = df['toothpaste']
plt.figure(figsize=(8, 5), dpi=100)
plt.plot(range(1, 13), data, 'o', markersize=8, label='ToothPaste Sales Data')
plt.grid(True, linestyle='--', linewidth=1.5)
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('ToothPaste Sales data')
plt.xticks(range(1, 13))
plt.legend(fontsize=12)
plt.show()
```

Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

```python
cream = df['facecream']
wash = df['facewash']

plt.figure(figsize=(8, 5), dpi=100)

plt.bar(df.index, cream, width=0.4, align='center', label='Face Cream sales data', color='red', alpha=0.7)
plt.bar(df.index, wash, width=0.4, align='edge', label='Face Wash sales data', color='blue', alpha=0.7)

plt.grid(True, linestyle='--', linewidth=1.5)
plt.xticks(range(1, 13))
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Facewash and Facecream sales data')
plt.legend(fontsize=12)

plt.show()
```

Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
data = df['total_profit']
plt.figure(figsize=(8, 5), dpi=100)

# Tính số lượng quan sát trong mỗi khoảng lợi nhuận
bin_counts, bin_edges, _ = plt.hist(data, bins=[150000, 175000, 200000, 225000, 250000, 300000, 350000])

# Định dạng số lượng quan sát hiển thị trên các cột
for count, edge in zip(bin_counts, bin_edges):
    plt.text(edge + 5000, count + 0.2, str(int(count)), ha='center', fontsize=9)

plt.xlabel('Profit range in dollar')
plt.ylabel('Frequency')
plt.title('Profit Distribution')
plt.grid(True, linestyle='--', linewidth=1.5)
plt.xticks(bin_edges)
plt.legend(labels=['Profit data'], loc='upper left', fontsize=11)

plt.show()
```

Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
langs = ['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
check = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
Sales = [df[column].sum() for column in check]

plt.figure(figsize=(8, 5), dpi=100)
plt.pie(Sales, labels=langs, autopct='%1.1f%%', startangle=90)
plt.title('Total Sales Data for Last Year by Product')
plt.legend(loc='lower right', fontsize=10)

plt.show()
```

Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
bathingsoap = df['bathingsoap']
facewash = df['facewash']

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), dpi=1000)

ax1.plot(bathingsoap, linewidth=3, marker='o', markersize=8, color="black")
ax1.set_title('Sales data of Bathingsoap')
ax1.set_xticks(range(1, 13))
ax1.set_yticks([7500, 10000, 12500])

ax2.plot(facewash, linewidth=3, marker='o', markersize=8, color="red")
ax2.set_title('Sales data of Facewash')
ax2.set_xticks(range(1, 13))
ax2.set_yticks([1500, 2000])

fig.suptitle('Sales data', fontsize=14)
fig.tight_layout()

plt.xlabel("Month Number")
plt.ylabel('Sales units in number')

plt.show()
```

Exercise 10: Read all product sales data and show it using the stack plot

```python
import numpy as np

plt.figure(figsize=(8, 4), dpi=1000)

data = np.array([df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer']])
colors = ['m', 'c', 'r', 'k', 'g', 'y']

plt.stackplot(df.index, data, labels=df.columns, colors=colors)
plt.title('All product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.legend(loc='upper left')

plt.show()
```
