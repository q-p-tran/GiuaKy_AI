```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv', sep='\t')
chipo.head(10)
```

Step 5. Create a histogram of the top 5 items bought

```python
import matplotlib.pyplot as plt

# Get the top 5 most ordered items
top_5_items = chipo['item_name'].value_counts().head(5)

plt.bar(top_5_items.index, top_5_items.values)

# Set the chart title and labels
plt.title('Top 5 Most Ordered Items')
plt.xlabel('Items')
plt.ylabel('Number of Orders')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

Step 6. Create a scatterplot with the number of items orderered per order price

```python
import matplotlib.pyplot as plt

# Convert the item_price column to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Group the orders by order_id and calculate the total order price and number of items
orders = chipo.groupby('order_id').agg({'item_price': 'sum', 'quantity': 'sum'})

# Create a scatterplot with order price on the x-axis and number of items on the y-axis
plt.scatter(orders['item_price'], orders['quantity'])

# Set the chart title and labels
plt.title('Number of Items Ordered per Order Price')
plt.xlabel('Order Price')
plt.ylabel('Number of Items Ordered')

# Display the chart
plt.show()
```
