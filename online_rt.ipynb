Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter out orders from UK
online_rt = online_rt.query("Country != 'United Kingdom'")

# Group the data by country and find the total quantity ordered
grouped = online_rt.groupby('Country')['Quantity'].sum()

# Sort the grouped data in descending order of quantity and select top 10 countries
top_10_countries = grouped.sort_values(ascending=False).head(10)

# Create a histogram of the top 10 countries
top_10_countries.plot(kind='bar')

# Set the title and axis labels
plt.title('Top 10 Countries by Quantity Ordered (except UK)')
plt.xlabel('Country')
plt.ylabel('Quantity Ordered')

# Display the histogram
plt.show()
```
